## 03 Solution part 1

```python
# 1.1 Iterate a list of names to return a list of the names starting with H
my_list = ["Eko", "EkoFIVE", "EkoTEN", "TEKO", "ealon"]
print(my_list)

filtered = [name for name in my_list if name[0].lower() == "e"] 
print(filtered)
```

```python
# 1.2 In one line create a list of the numbers 1-100 to the power of 3

my_secList = [element**3 for element in list(range(0,100))]
print(my_secList)
```

```python
# 1.3 Iterate a list of names to create a list of tuples where the tuples first value is the length of 
# the name and the second is the name

my_list = ["Eko", "EkoFIVE", "EkoTEN", "TEKO", "ealon"]

my_modifiedList = [(len(element), element) for element in my_list]
print(my_modifiedList)
```

```python
# 1.4 Iterate over each character in a string and get only those that are nummeric

nstr = "aa01w5s0ettE77a"

my_numericList = [element for element in nstr if element.isnumeric()]
print(my_numericList)
```

```python
# 1.4 Using only a list comprehension wrapped in set() get all possible combination from throwing 2 dice 
# (hint use 2 for loops in a single list comprehension)

diceOne = [1, 2, 3, 4, 5, 6]
diceTwo = [1, 2, 3, 4, 5, 6]

combos = [str(d1) +', ' + str(d2) for d1 in diceOne for d2 in diceTwo]
print(combos)
```
